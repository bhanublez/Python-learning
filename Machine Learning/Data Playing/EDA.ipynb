## What is EXploratory Data Analysis (EDA)
We know that it is an important concept of ML projects. Here major role is to dig into the data to understand more about the data.
1. Know the statistic of data
2. Spot and handle missing values in data
3. balance imbalance data 
4. Relations between different different feature
5. And understanding too this feature especially the important ones
