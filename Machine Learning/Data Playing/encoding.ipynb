```python
titanic=sns.load_dataset('titanic')
titanic_c=titanic.dropna()#drop all missing value
```

```python
titanic_c.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

deck_feat = titanic_c[['deck']]

label_encoder = LabelEncoder()

deck_encoded = label_encoder.fit_transform(deck_feat)
```

```python
titanic_c['deck'] = pd.DataFrame(deck_encoded, columns=deck_feat.columns, index=deck_feat.index)

titanic_c.head()
```

```python
titanic_c['deck'].value_counts()
```
