# ***Matplotlib***

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
#simple liner plotting
x=[1,2,3,4,5,6,7,8,9]
y=np.exp(x)
plt.plot(x,y)
plt.title('Exp linear Graph')
plt.xlabel('x data on x axis')
plt.ylabel('y data on y axis')
````

## Subplots

````python
plt.subplot(1,3,1)
plt.plot(x,y)

plt.subplot(1,3,2)
plt.plot(x,np.sin(y))

plt.subplot(1,3,3)
plt.plot(x,np.cos(y))
````

````python
plt.figure(figsize=(10,8))#plt.figure(figsize=(width,height))
plt.plot(x,y)
````

````python
plt.figure(figsize=(8,6),dpi=100)#dpi stand for dot per inch
plt.plot(x,y)
````

````python
figure=plt.figure(figsize=(8,5))
plt.plot(x,y)
figure.savefig('myfigure.png')
````

````python
x=np.linspace(0,50,100)#from 0 coordinate to (50,100) position

a=np.sin(x)
plt.title('Sine Graph')
plt.plot(x,a)
````

````python
b=np.cos(x)
plt.title('Cos Graph')
plt.plot(x,b)
````

````python
c=np.tan(x)
plt.title('Tan Graph')
plt.plot(x,c)
````

````python
plt.plot(x,a,label='Sin function')
plt.plot(x,b,label='Cos function')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend(loc='upper right')
````

````python
plt.figure(figsize=(10,8))
plt.plot(x,a,color='red')
plt.plot(x,b,color='y')
plt.plot(x,b*2,color='green')
plt.plot(x,b*3,color='0.8')
````

### Line Styles and marker

```
'-' ....solid line style
'--'....dashed line style
'-.'....dash-dot line style\
':'....dotted line style
'b'...# blue markers with default shape
'or'...# red circles
'-g'....# green solid line
'--'...# dashed line with default color
'^k:'...# black triangle_up markers connected by a dotted line

```

```
Line Markers


'.'...point marker
','...pixel marker
'o'...circle marker
'v'...triangle_down marker
'^'...triangle_up marker
'<'....triangle_left marker
'>'....triangle_right marker
'1'....tri_down marker
'2'....tri_up marker
'3'....tri_left marker
'4'....tri_right marker
's'....square marker
'p'....pentagon marker
'*'....star marker
'h'....hexagon1 marker
'H'....hexagon2 marker
'+'....plus marker
'x'....x marker
'D'....diamond marker
'd'....thin_diamond marker
'|'....vline marker
'_'....hline marker
```

````python
plt.figure(figsize=(10,8))
plt.plot(x,a,color='red',ls='-',marker='+',lw=1,markersize=10)
plt.plot(x,b,color='pink',ls='--',marker='o',lw=1,markersize=10)
plt.plot(x,b*2,color='purple',ls='-.',marker='s',lw=1,markersize=10)
````

````python
x=[1,2,3,4,5,6,7,8]
y=np.square(x)
plt.plot(x,y)
plt.text(1,50,'Bhai yeah text 1,50 coordinate per ayaa',color='brown')
````

````python
plt.plot(x,y)

plt.annotate('This is the Minimum Point',xy=(1,1),xytext=(2,25),arrowprops=dict(facecolor='r'))

plt.annotate('This is the max point',xy=(8,64),xytext=(1,60),bbox=dict(boxstyle='round'),arrowprops=dict(arrowstyle='fancy',fc='Green'))

plt.annotate('This is mid area',xy=(4.5,22),xytext=(5,12),bbox=dict(boxstyle='round',fc='none'), arrowprops=dict(arrowstyle='wedge, tail_width=0.5', alpha=0.3))
````

````python
fig=plt.figure()

ax=plt.axes()#axes

ax.plot(x,y,'y')
ax.set_title=('Kuch Toh Title hai')
ax.set_xlabel('X label')
ax.set_ylabel('Y Label')
````

````python
fig,axes=plt.subplots(1,2)#(row,columns)
axes[0].plot(x,y,'r')
axes[1].plot(x,y,'b')
axes[0].set_title('Sub plot 1')
axes[1].set_title('Sub plot 2')
````

````python
fig,axes=plt.subplots(1,3)
j=0
for i in axes:
    i.plot(x,y,'g')
    j+=1
    i.set_title(f'This is {j} title')

plt.tight_layout()

# fig
````

````python
ax=plt.axes()
ax.plot(x,y,'g')
ax.set(title='This is a Title',ylim=(-10,10),xlim=(0,25))
````
