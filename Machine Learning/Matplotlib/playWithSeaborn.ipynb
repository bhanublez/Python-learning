# Play With Seaborn

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data set to be used
titanic = sns.load_dataset('titanic')
fmri=sns.load_dataset('fmri')
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
iris=sns.load_dataset('iris')
```

```python
#Let try to make relationship between these numeric features
sns.scatterplot(x='age',y='fare',data=titanic)
```

```python
sns.scatterplot(x='age',y='fare',hue='sex',data=titanic)#hue is used to differentiate between the data
```

```python
sns.scatterplot(x='age',y='fare',data=titanic,hue='sex',style='sex')
```

```python
plt.figure(figsize=(24,20))
sns.scatterplot(data=titanic,x='age',y='fare',hue='sex',style='pclass')
```

```python
plt.figure(figsize=(12,10))
markers=['o','v','s']
sns.scatterplot(data=titanic,x='age',y='fare',hue='sex',style='pclass',markers=markers ,s=100)
# sns.scatterplot(data=titanic, x='age', y='fare', hue='sex', size='sex', sizes=(20,200))
```

```python
#Lets Try to make a line plot
sns.lineplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.regplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.relplot(data=flights, x='month', y='passengers', kind='line')
```

```python
sns.relplot(data=flights, x='year', y='passengers', kind='line')
```

```python
sns.set_theme(style='darkgrid')
sns.relplot(data=flights, x='year', y='passengers', kind='line')
```

```python
sns.relplot(data=flights, x='year', y='passengers', kind='line',hue='month')
```

```python
sns.relplot(data=flights, x='month', y='passengers', ci=None, estimator=None, kind='line')
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', 
            hue='event', kind='line',style='event',markers=True)
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', hue='event',kind='line',style='event',dashes=False,markers=True)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',col='time',data=tips)
```
