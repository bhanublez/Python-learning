```python
train_input_data = mob_train.drop('price_range', axis=1)
train_labels = mob_train['price_range']
```

```python
from sklearn.preprocessing import MinMaxScaler #Normalization
scaler=MinMaxScaler()
scaler.fit(train_input_data)
train_input_data=scaler.transform(train_input_data)
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
log_model.fit(train_input_data,train_labels)
```

```python
log_model.score(train_input_data,train_labels)
```

```python
from sklearn.linear_model import SGDClassifier
sgd_model=SGDClassifier()
sgd_model.fit(train_input_data,train_labels)
```

```python
sgd_model.score(train_input_data,train_labels)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model=DecisionTreeClassifier()
tree_model.fit(train_input_data,train_labels)
```

```python
tree_model.score(train_input_data,train_labels)
```

```python
#Model Evaluation
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import cross_val_score
log_score=cross_val_score(log_model,train_input_data,train_labels,cv=10)
```

```python
log_score
```

```python
log_score.mean()
```

```python
from sklearn.model_selection import cross_val_score
sgd_scores=cross_val_score(sgd_model,train_input_data,train_labels,cv=10)
```

```python
sgd_scores
```

```python
sgd_scores.mean()
```

```python
from sklearn.model_selection import cross_val_score
tree_scores=cross_val_score(tree_model,train_input_data,train_labels,cv=10)
```

```python
tree_scores
```

```python
tree_scores.mean()
```

```python
#cross validation we get that tree method also not good so we consider logistic regression now apply prediction on test data
from sklearn.model_selection import cross_val_predict
predict_on_train=cross_val_predict(log_model,train_input_data,train_labels,cv=10)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix=confusion_matrix(train_labels,predict_on_train)
```

```python
confusion_matrix
```

```python
sns.heatmap(confusion_matrix,annot=True,cmap='PuBu')
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(train_labels,predict_on_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(train_labels,predict_on_train))
```
